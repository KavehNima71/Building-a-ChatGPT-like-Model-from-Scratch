### üü¢ Comparison of Outputs:


| Library  | Output                                                                 |
|----------|------------------------------------------------------------------------|
| `spaCy`  | `['Mr.', 'Smith', 'bought', 'cheapsite.com', 'for', '$1.5', 'million', 'in', '2023', '.']` |
| `nltk`   | `['Mr.', 'Smith', 'bought', 'cheapsite.com', 'for', '$', '1.5', 'million', 'in', '2023', '.']` |
| `gensim` | `['mr', 'smith', 'bought', 'cheapsite', 'com', 'for', 'million', 'in']` |

### üü¢ Analysis of Outputs:

1. **`spaCy`**:
   - Handles abbreviations (`Mr.`), domain names (`cheapsite.com`), and currency (`$1.5`) well.
   - Produces the most accurate and clean tokens.

2. **`nltk`**:
   - Handles abbreviations and domain names well but splits currency into `$` and `1.5`.
   - Still a robust tokenizer for most use cases.

3. **`gensim`**:
   - Simplifies tokens by lowercasing and removing punctuation.
   - Loses some important information (e.g., `Mr.` becomes `mr`, `$1.5` is lost).
   - Best suited for preprocessing text for topic modeling or similarity tasks.

### üü¢ When to Use Which Library:


- Use **`spaCy`** for production-grade NLP tasks where accuracy is critical.
- Use **`nltk`** for educational purposes or when you need a lightweight tokenizer.
- Use **`gensim`** for tasks like topic modeling or document similarity, where simple preprocessing is sufficient.

## üü† Train a BPE Tokenizer (Simple Example)

```python
# Example training data (in practice, use a larger dataset)
ds = [
    "low lower new newest widest",
    "low lower new new wider",
    "lowest wider wider newest",
]
```

```python
from tokenizers import Tokenizer, models, pre_tokenizers, trainers
```

```python
# 1. Create an empty BPE tokenizer
tokenizer = Tokenizer(models.BPE(unk_token="[UNK]"))
tokenizer

# 2. Use a pre-tokenizer to split words based on whitespace
# tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

# 3. Define a trainer for training the tokenizer
trainer = trainers.BpeTrainer(
    vocab_size=40,  # Set the vocabulary size
    special_tokens=["[UNK]"],
    min_frequency=2,  # Set the minimum frequency of tokens
)

# 4. Train the tokenizer on the dataset
tokenizer.train_from_iterator(ds, trainer)

# 5. Save the trained tokenizer
tokenizer.save("bpe_tokenizer.json")

# 6. Test the tokenizer
tokenizer.encode("below lowest new").tokens
# tokenizer
```

```python
vocab = tokenizer.get_vocab()
sorted_vocab = dict(sorted(vocab.items(), key=lambda item: item[1]))  # Sort by token IDs
print(sorted_vocab)
```

## üü† Train a Custom Tokenizer

You can also train your own tokenizer on a custom dataset. Here's an example of training a **Byte Pair Encoding (BPE)** tokenizer:

```python
dataset["validation"]
```

```python
from tokenizers import Tokenizer, normalizers, pre_tokenizers, models, processors, decoders, trainers
```

```python
# Initialize a BPE tokenizer
tokenizer = Tokenizer(models.BPE(unk_token="[UNK]"))

# Use a pre-tokenizer to split text into words
tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel()

# Initialize a BPE trainer
trainer = trainers.BpeTrainer(
    vocab_size=10000,  # Set the vocabulary size
    special_tokens=["[UNK]", "<|endoftext|>"],
    min_frequency=2,  # Set the minimum frequency of tokens
    )

# Train the tokenizer on a custom dataset
tokenizer.train_from_iterator(dataset["validation"]["text"], trainer)
print(tokenizer.get_vocab_size())

# Add special tokens
tokenizer.post_processor = processors.TemplateProcessing(
    single="<|endoftext|> $A <|endoftext|>",
    special_tokens=[("<|endoftext|>", tokenizer.token_to_id("<|endoftext|>"))],
)

# Add decoder
tokenizer.decoder = decoders.ByteLevel()

# Tokenize a sample text
text = "Mr. Smith bought cheapsite.com for $1.5 million in 2023."
output = tokenizer.encode(text)

print("Tokens:", output.tokens)
print("Input IDs:", output.ids)
```

```python
# Load the GPT-2 tokenizer for both libraries
tiktokenizer = tiktoken.get_encoding("gpt2")  # tiktoken

# Encode text (returns token IDs)
token_ids = tiktokenizer.encode(sample)

# Convert token IDs to actual tokens
tokens = [tiktokenizer.decode_single_token_bytes(tid).decode("utf-8", errors="ignore") for tid in token_ids]

# Decode token IDs
decoded_text = tiktokenizer.decode(token_ids)

# Print results
pprint(sample)
print(100*"-")

print(token_ids)
print(100*"-")

print(tokens)
print(100*"-")

pprint(decoded_text)
```
